#### Remove High Variable Correlation

```python
columns = df[['tenure','ServiceCount','MonthlyCharges','TotalCharges']]
columns
```

```python
columns.corr()
```

```python
# sns.heatmap(columns.corr(), annot=True, vmin=-1, vmax=1)
sns.heatmap(columns.corr(), annot=True)
```

```python
lower = pd.DataFrame(np.tril(columns.corr(), -1), columns=columns.corr().columns)
lower
```

#### Remove Variable with High p-value from Chi-square

```python
columns = df[['Churn']]
columns
```

```python
columns['Churn'].unique()
```

```python
columns['Churn'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
# variables = df[['gender', 'Contract', 'PaperlessBilling']]
columns['Churn'] = label_encoder.fit_transform(columns['Churn'])
columns['gender'] = label_encoder.fit_transform(df['gender'])
columns['Contract'] = label_encoder.fit_transform(df['Contract'])
columns['PaperlessBilling'] = label_encoder.fit_transform(df['PaperlessBilling'])
```

```python
columns
```

```python
outputs = columns[['Churn']]
variables = columns[['gender', 'Contract', 'PaperlessBilling']]
```

```python
outputs
```

```python
variables
```

```python
from sklearn.feature_selection import chi2
chi_table = chi2(variables, outputs)
chi_table
```

```python
p_value = chi_table[1]
p_value = pd.DataFrame(np.tril(p_value, -1), columns=variables.columns)
p_value
```

```python
to_drop = [column for column in p_value if any(p_value[column] > 0.05)]
to_drop
```

```python
variables.drop(to_drop, inplace=True, axis=1)
```

```python
variables
```

```python
df.drop(labels=['Churn', 'gender', 'Contract', 'PaperlessBilling'], axis=1, inplace=True)
df = df.join([variables, outputs])
df
```
