```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc = {'figure.figsize':(15,8)})
sns.set_theme(style="whitegrid")
```

## Data Explore

```python
df = pd.read_csv('CarPrice.csv')
```

```python
df.info()
```

```python
df.drop(['car_ID', 'CarName'], axis=1, inplace=True)
df
```

```python
categorical_features = df.select_dtypes(include=['object'])
df = df.select_dtypes(exclude=['object'])
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.loc[:, df.columns != 'price']
X
```

```python
Y = df['price']
```

## Data Transfrom and Feature Selection

```python
X_standard = pd.DataFrame()
X_standard[X.columns] = StandardScaler().fit_transform(X)
X_standard
```

```python
X_standard.corr()
```

```python
sns.heatmap(X_standard.corr(), annot=True)
```

```python
lower = pd.DataFrame(np.tril(X_standard.corr(), -1), columns=X_standard.corr().columns)
lower
```

```python
sns.heatmap(lower, annot=True)
```

```python
to_drop = [column for column in lower if any(lower[column] > 0.86)]
to_drop
```

```python
X_standard.drop(to_drop, inplace=True, axis=1)
X_standard
```

```python
categorical_features
```

```python
df_encoded = pd.get_dummies(data=categorical_features, drop_first=True)
df_encoded
```

```python
X_standard[df_encoded.columns] = df_encoded
X_standard
```

## PCA

```python
pca = PCA()
X_pca = pca.fit_transform(X_standard)
```

```python
print('Explained Variznce ratio = ', pca.explained_variance_ratio_)
```

```python
plt.bar(np.arange(1, len(pca.explained_variance_ratio_) + 1), height=pca.explained_variance_ratio_) 
plt.show()
```

```python
print('Explained Variance (eigenvalues) = ', pca.explained_variance_)
print('-----------------------------')
print('PCA components (eigenvectors)')
print(pca.components_[0:2, :])
```

#### PCA 2

```python
pca2 = PCA(n_components=5)
X_pca_2 = pca2.fit_transform(X_standard)
```

```python
print('Explained Variznce ratio = ', pca2.explained_variance_ratio_)
```

```python
plt.margins(x=0.5)   
plt.bar(np.arange(1, 6), height=pca2.explained_variance_ratio_) 
```

```python
print('Explained Variance (eigenvalues) = ', pca2.explained_variance_)
print('-----------------------------')
print('PCA components (eigenvectors)')
print(pca2.components_[0:2, :])
```

## Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
Rseed = 100
x_train_set, x_test, y_train_set, y_test = train_test_split(X_pca_2, Y, test_size = 0.3, random_state = Rseed)
```

```python
x_train, x_validate, y_train, y_validate = train_test_split(x_train_set, y_train_set, test_size = 0.3, random_state = Rseed)
```

```python
# Perform Linear Regression -> All variables
lr = LinearRegression()
# Train
lr.fit(x_train, y_train)
# Validate
y_pred_lr = lr.predict(x_validate)
# Test
y_test_pred_lr = lr.predict(x_test)
```

```python
# Measure Accuracy Validation and Test
print(r2_score(y_pred_lr, y_validate))
print(r2_score(y_test_pred_lr, y_test))
print(lr.score(x_validate, y_validate))
print(lr.score(x_test, y_test))
```
